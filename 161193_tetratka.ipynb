Овој научен труд сака да ја покаже врската меѓу типот на личност и типот на играч/гејмер. Презентира еден модел наречен BrainHex, кој според анкета ви покажува каков тип на гејмер сте. Има 7 типа на гејмери: Seeker, Survivor, Daredevil, Mastermind, Conqueror, Socialiser, Achiver. За да ова истражување биде изводливо, треба да си го знаете вашиот тип на личност (Personality). Според тоа каква личност сте (дали екстровертна, интровертна, се водите според чувства итн.) и каков тип на гејмер сте (според анкетата), гледа дали постои некаква поврзаност меѓу нив. Испитани се околу 50.000 поединци кој во поголем број се од машки пол, и остава простор за понатамошно истражување.
